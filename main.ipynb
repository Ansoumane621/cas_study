```python
import shutil, os

base = "/content/drive/MyDrive/Colab Notebooks/genIA - Copie"
space_dir = "/content/space_upload"
os.makedirs(space_dir, exist_ok=True)
os.makedirs(f"{space_dir}/src", exist_ok=True)
os.makedirs(f"{space_dir}/data", exist_ok=True)

shutil.copy(f"{base}/hf_space/app.py", f"{space_dir}/app.py")
shutil.copy(f"{base}/src/context_state_machine.py", f"{space_dir}/src/context_state_machine.py")
shutil.copy(f"{base}/src/data_utils.py", f"{space_dir}/src/data_utils.py")
shutil.copy(f"{base}/data/syntetic_wolof_instruct_data.jsonl", f"{space_dir}/data/syntetic_wolof_instruct_data.jsonl")

open(f"{space_dir}/src/__init__.py", "w").close()

with open(f"{space_dir}/requirements.txt", "w") as f:
    f.write("torch\ntransformers\npeft\naccelerate\ngradio\n")

print("ready:", os.listdir(space_dir))
```
